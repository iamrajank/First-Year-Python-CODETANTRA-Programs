```python
fin = open('InputData2.txt' , 'r')
charCount = wordCount = lineCount = 0			#Initialize Counters
for line in fin:								#Read each Line

	# write your logic here
    line.strip("\n")
    words=line.split()
    
	#Increment Line count
    lineCount+=1
    
	#split() gives the words in a list
    wordCount+=len(words)
    
	#Increment Character Count
    charCount+=len(line)
fin.close()

print("Line count = ", lineCount)  #Print the Counts
print("Word count = ", wordCount)
print("Char count = ", charCount)
```

```python
li1=input("Enter the list of numbers: ").split()
li=[int(i) for i in li1]
def bubble(li):
	for i in range(len(li)-1):
		for j in range(0,len(li)-i-1):
			if li[j]>li[j+1]:
				li[j],li[j+1]=li[j+1],li[j]
bubble(li)
print(f'Sorted list: {li}')
```
